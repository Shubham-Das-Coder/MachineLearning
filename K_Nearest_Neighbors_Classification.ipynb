```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris.feature_names #Getting the feature names
```

```python
iris.target_names #Getting the types of flowers by name
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names) #loading iris.data in dataframe
df.head() #Getting the first five rows
```

```python
df.shape #Getting the shape of the dataframe
```

```python
df['target']=iris.target #Adding a new column named target
df.head() #Displaying the first five rows
```

```python
df[df.target==0] #Displaying all the rows where target=0(ie,setosa)
```

```python
df[df.target==1] #Displaying all the rows where target=1(ie,versicolor)
```

```python
df[df.target==2] #Displaying all the rows where target=2(ie,verginica)
```

```python
#Splitting the dataframe into three separate dataframes
df0=df[:50]     #Setosa flowers
df1=df[50:100]  #Versicolor flowers
df2=df[100:150] #Virginica flowers
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

SEPAL LENGTH VS SEPAL WIDTH (SETOSA VS VERSICOLOR)

```python
plt.xlabel('Sepal Length') #Label of x-axis
plt.ylabel('Sepal Width')  #Label of y-axis
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color="green",marker="+") #Setosa flowers
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color="blue",marker=".")  #Versicolor flowers
```

PETAL LENGTH VS PETAL WIDTH (SETOSA VS VERSICOLOR)

```python
plt.xlabel('Petal Length') #Label of x-axis
plt.ylabel('Petal Width')  #Label of y-axis
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color="green",marker="+") #Setosa flowers
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color="blue",marker=".")  #Versicolor flowers
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['target'],axis='columns') 
y=df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

CREATE KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3) #The value of k is 3
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10) #The value of k is 10
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred=knn.predict(X_test)

cm=confusion_matrix(y_test,y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,5))
sns.heatmap(cm,annot=True)
plt.xlabel=('Predicted')
plt.ylabel=('Truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
