```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("GRAPH OF SINE FUNCTION")
plt.plot(x,y,'b')
```

```python
x=np.arange(0,5*np.pi,0.1)
ysin=np.sin(x)
ycos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,ysin)
plt.title("SINE")
plt.subplot(2,1,2)
plt.plot(x,ycos)
plt.title("COSINE")
```
