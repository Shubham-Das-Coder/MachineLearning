```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Suppress warnings for claen notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read dataset
dataset=pd.read_csv("melb_data.csv")
dataset
```

```python
dataset.nunique() #Displaying number of unique things
```

```python
dataset.shape #Displaying the shape of the dataset
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount',
            'Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset=dataset[cols_to_use]
dataset
```

```python
dataset.shape
```

```python
dataset.isna().sum() #Values are NaN
```

```python
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car'] #Filling these columns default value as zero
dataset[cols_to_fill_zero]=dataset[cols_to_fill_zero].fillna(0) #Filling value as zero where value is NaN
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
X=dataset.drop('Price',axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.9,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
